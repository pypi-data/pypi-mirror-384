```python
import secrets

def uniform(a: float = 0.0, b: float = 1.0) -> float:
    """Cryptographically secure uniform sample."""
    # 53 random bits gives 53-bit precision double
    u = secrets.randbits(53) / (1 << 53)  # in [0, 1)
    return a + (b - a) * u
```

```python
x = []
for i in range(0, 2000):
    x.append(uniform(2, 5))
```
