# Text Snippet Guide

Text Snippets allow you to create, format and finally upload text to a snip lab book.

```python
from snip.snippets import TextSnip

text = TextSnip("Hello, World!")
text
```

As with any snippet you may change the position of the snippet by modifying the `x` and `y` attributes.

```python
text.x = 400
text.y = 1000
text.rotation = 45
text
```

You may also get the width and height of the snippet, but you may not change them. Similarly, you may get the bounding box of the snippet.

```python
bounds = text.bounding_box
print(f"Height: {text.height}, Width: {text.width}")
print(f"Bounds: {bounds}")
```

## Text Styling

You may apply styles to text snippets by using the available attributes of the [TextSnip](snip.snippets.text.TextSnip) class or the `_style` attribute. For instance to change to font size of a text snippet you may use the [font_size](snip.snippets.text.TextSnip.font_size) attribute.


```python
text.font_size = 50
text
```

By default lines wrap automatically at 400 pixels, you may change this by setting the [line_wrap](snip.snippets.text.TextSnip.line_wrap) attribute. You may disable text wrap by setting the line_wrap attribute to `-1`!

```python
text.font_size = 30
text.line_height = 1.5
text.text = "I should wrap once the text reaches 400 pixels!"
text
```

Again you may upload the snippet by calling the [upload](snip.snippets.base.BaseSnip.upload) method which should be inherited by all snippets.

```python
text.upload(book_id=1)
```
