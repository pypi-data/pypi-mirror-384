Generate ShEx shapes from VoID descriptiom

```python
from sparql_llm.sparql_void_shapes_loader import (
    get_shex_dict_from_void,
    get_shex_from_void,
)

shex_dict = get_shex_dict_from_void("https://sparql.uniprot.org/sparql/")
print(len(shex_dict))
for cls, shex_shape in shex_dict.items():
    print(cls)
    if "label" in shex_shape:
        print(shex_shape["label"])
    if "comment" in shex_shape:
        print(shex_shape["comment"])
    print(shex_shape["shex"])
    print()
```
