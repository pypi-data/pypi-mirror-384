# Example Usage of the FlipPartnerMap class

This class is useful for verifying that a flip sequence is actually valid. It also provides functionality to build your algorithm on top of it, e.g., by searching for flippable edges or applying flips.

```python
from cgshop2026_pyutils.geometry import FlippableTriangulation, draw_flips, Point, draw_edges

points = [Point(0, i) for i in range(4)] + [Point(1, i) for i in range(4)]
edges = [(0, 5), (1,5), (1, 6), (2,6), (2,7), (3,7)]
draw_edges(points, edges, show_indices=True)
```

```python
triangulation: FlippableTriangulation = FlippableTriangulation.from_points_edges(points, edges)
draw_flips(triangulation, show_indices=True)
```

```python
triangulation.possible_flips()  # list of flippable edges
```

```python
triangulation.add_flip((0, 5))
draw_flips(triangulation, show_indices=True)
```

```python
triangulation.possible_flips()  # list of flippable edges
```

```python
triangulation.add_flip((1, 6))
draw_flips(triangulation, show_indices=True)
```

```python
triangulation.possible_flips()  # list of flippable edges
```

```python
triangulation.add_flip((2, 6))
```

```python
triangulation.commit()
draw_flips(triangulation, show_indices=True)
```

```python
triangulation.possible_flips()  # list of flippable edges
```
