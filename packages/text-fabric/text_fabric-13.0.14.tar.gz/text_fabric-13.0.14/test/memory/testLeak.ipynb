```python
import gc
from tf.app import use
from tf.fabric import Fabric
```

```python
gc.set_debug(gc.DEBUG_LEAK)
```

```python
TF = Fabric(locations="~/text-fabric-data/etcbc/bhsa/tf", modules="c")

dest = {}

def loadIt(i):
    api = TF.load('g_word_utf8')
    dest[i] = api
    
for i in range(3):
    print(i)
    loadIt(i)
```

```python
dest = {}

def loadIt(i):
    A = use('bhsa')
    dest[i] = A
    
for i in range(3):
    print(i)
    loadIt(i)
```

```python
for (i, A) in dest.items():
    print(f"{i} => API object                      is {hex(id(A.api))}")
print()
for (i, api) in dest.items():
    print(f"{i} => g_word_utf8 feature object      is {hex(id(A.api.F.g_word_utf8))}")
print()
for (i, api) in dest.items():
    print(f"{i} => g_word_utf8 feature data object is {hex(id(A.api.F.g_word_utf8.data))}")
```
