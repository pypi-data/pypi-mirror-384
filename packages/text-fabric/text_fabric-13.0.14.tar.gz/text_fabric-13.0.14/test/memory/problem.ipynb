# Problem

Find the best memory layout for feature data.

A feature is a mapping from natural numbers to values.
The values are either all integers, or all strings

All features are read only.

A typical dataset has 50-100 features, mapping 1-2 million integers to values.

But:

* there are very small features, that only map a few or a few thousand numbers
* the bulk of the features only map the numbers 1 to max, for a fixed max, typically 500,000 or so
* some features map many nodes to only a few discrete values
* some features have many holes in the mapped integers

Criteria:

* loading feature data should be fast
* reading feature data should be fast
* the memory footprint should be minimal

Examples:

`otype` = 

```
1-500000 word
500000-700000 phrase
700000-800000 clause
800000-850000 sentence
850000-875000 verse
875000-880000 chapter
880000-880050 book
```

`partofspeech` = 

```
art
noun
verb
art
verb
adj
noun
pronoun
adj
noun
...
```

`tense` = 

```
3 past
5 past
10 present
16 future
18 present
```

`annotation` = 

```
350000 this
350001 is
350002 your
350003 annotation

450000 this
450001 is
450002 my
450003 annotation
```

## Solutions

A python dict is a data structure that can store features of this kind rather efficiently.
