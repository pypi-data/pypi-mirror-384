```python
from tf.server.kernel import makeTfConnection
from tf.applib.appmake import findAppConfig
from tf.core.helpers import console
```

```python
config = findAppConfig('bhsa')
config
```

```python
TF = makeTfConnection(config.host, config.port)
kernelApi = TF.connect()
```

```python
querySlow = '''
p:phrase
    =: wFirst:word
    wLast:word
    :=

wGap:word
wFirst < wGap
wLast > wGap

p || wGap

v:verse

v [[ wFirst
v [[ wGap
'''

queryFast = '''
verse
    p:phrase
      wPreGap:word lex=L
      wLast:word
      :=

wGap:word
wPreGap <: wGap
wGap < wLast
p || wGap
'''
```

```python
def search(query):
    (results, messages) = kernelApi.rawSearch(query)
    if messages:
        console(messages, error=True)
    else:
        console(f'{len(results)} results')
    return results
```

```python
def search(query):
    results = kernelApi.rawSearch(query)
    console(f'{len(results)} results')
    return results
```

```python
search(queryFast)
```

```python
search(querySlow)
```
