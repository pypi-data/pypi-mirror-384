```python
from tf.server.provider import TfProvider
```

```python
TfP = TfProvider('nena', checkoutApp='local')
```

# Passage reduced to slots

```python
TfP.passage('Barwar', 'A Hundred Gold Coins', 1)
```

# Passage reduced to morphemes

```python
TfP.passage('Barwar', 'A Hundred Gold Coins', 1, baseTypes={'morpheme'})
```

# Passage reduced to words

```python
TfP.passage('Barwar', 'A Hundred Gold Coins', 1, baseTypes={'word'})
```

# Passage reduced to words and then morphemes

```python
TfP.passage('Barwar', 'A Hundred Gold Coins', 1, baseTypes={'morpheme', 'word'})
```

# Passage reduced to words and then morphemes and then letters

```python
TfP.passage('Barwar', 'A Hundred Gold Coins', 1, baseTypes={'morpheme', 'word', 'letter'})
```

# Passage reduced to morphemes with feature foreign included

```python
TfP.passage('Barwar', 'Gozali and Nozali', 29, baseTypes={'morpheme'}, features='foreign')
```
