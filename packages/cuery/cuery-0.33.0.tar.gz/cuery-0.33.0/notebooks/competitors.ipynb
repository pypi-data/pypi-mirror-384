```python
import time

importlib.reload(brands)

params = {
    "brand": "Coca-Cola",
    "sector": "Bebidas",
    "market": "España",
    "instructions": "Focus on soft drink multinationals, not individual products/drinks or local brands.",
    "search_kwargs": {"context_size": "medium", "reasoning_effort": "low"},
}

cfg = brands.BrandSearchConfig(**params)

start_wall = time.perf_counter()  # wall clock time
start_cpu = time.process_time()
infos = await brands.search_brands(cfg, to_pandas=True)
end_wall = time.perf_counter()
end_cpu = time.process_time()

print(f"Wall time: {end_wall - start_wall:.3f} s")
print(f"CPU time:  {end_cpu - start_cpu:.3f} s")
infos
```
